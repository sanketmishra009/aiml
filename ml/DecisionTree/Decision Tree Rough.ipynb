```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
x = np.linspace(.0010 , 0.9999999, 1000)

def entropy(x):
    return  -(x * np.log2(x) + (1-x) * np.log2(1-x))

y = entropy(x)
```

```python
plt.plot(x , y)
```

```python

# import pickle
# Load data (deserialize)
with open('../data/preprocessed_X_sm.pickle', 'rb') as handle:
    # X_train = pickle.load(handle)
    xtrain = pd.read_pickle("../data/preprocessed_X_sm.pickle")

with open('../data/X_test.pickle', 'rb') as handle:
    # X_test = pickle.load(handle)
    xtest = pd.read_pickle("../data/X_test.pickle")


with open('../data/y_sm.pickle', 'rb') as handle:
    # y_train = pickle.load(handle)
    ytrain = pd.read_pickle("../data/y_sm.pickle")


with open('../data/y_test.pickle', 'rb') as handle:
    # y_test = pickle.load(handle)
    ytest = pd.read_pickle("../data/y_test.pickle")
```

```python
xtrain.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion = 'entropy')
model.fit(xtrain , ytrain)
```

```python
model.score(xtest , ytest)
```

```python
model.predict(xtest)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(14,14))
plot_tree(model, filled = True);
```

```python
model.feature_importances_
```

```python
indices = np.argsort(model.feature_importances_)[::-1]
names = [xtrain.columns[i] for i in indices]
```

```python
plt.figure(figsize = (20 , 10))
sns.barplot(x = names , y = np.sort(model.feature_importances_)[::-1] ,  palette = 'Blues')
plt.xticks(rotation = 90)
plt.show()
```
