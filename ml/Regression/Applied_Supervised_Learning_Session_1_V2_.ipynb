### **Why Handling Missing Values Is Important in Supervised Learning**

In supervised learning, models learn patterns from data. If data is missing (e.g., blank entries), the model might make wrong assumptions or even crash.

#### **Industry Example:**

A bank uses customer data (age, income, credit score) to decide who gets a loan.
If income data is missing for some customers and ignored, the bank might approve loans for people with unknown or very low incomes‚Äîleading to financial risk.

#### ** Data Example:**

```python
# Dummy dataset: Predict if customer will buy insurance
# Missing values (None) in 'income'
data = [
    {'age': 25, 'income': 50000, 'buy': 1},
    {'age': 40, 'income': None, 'buy': 0},
    {'age': 30, 'income': 70000, 'buy': 1},
]
```

If we don‚Äôt fill in or handle the missing income, the model could misinterpret these customers, leading to **unreliable predictions**.

---

**Bottom line:**

> **Handling missing values ensures your model learns the real patterns and avoids costly mistakes in the real world.**

#### üîÅ Imputation Methods

| Method                | When to Use                      |
| --------------------- | -------------------------------- |
| Mean/Median           | For numerical data               |
| Mode                  | For categorical data             |
| Predictive Imputation | Use another ML model to estimate |


````python
# Industry Example: Insurance Industry (Predicting Premiums)
# Insurance companies often predict customer premiums based on data like age, gender, and occupation. Missing data is common, especially for attributes like customer income. Let's see how to handle these missing values practically:

# Step 1: Creating Dummy Data
````

````python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
# Sample customer data with missing incomes
data = {
    'Age': [25, 30, 45, 40, 55, 50],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F'],
    'Occupation': ['Engineer', 'Doctor', 'Doctor', 'Engineer', 'Lawyer', 'Lawyer'],
    'Income': [50000, np.nan, 80000, np.nan, 120000, 100000]
}

df = pd.DataFrame(data)
print("Original Data:\n")
df
````

# Imputation using mean


````python
mean_income = df['Income'].mean()
df_mean = df.copy()
df_mean['Income'].fillna(mean_income, inplace=True)
print("\nMean Imputed:",mean_income)
df_mean
````

````python
median_income = df['Income'].median()
df_median = df.copy()
df_median['Income'].fillna(median_income, inplace=True)
print("\nMedian Imputed:", median_income)
df_median
````

# Using Groupby


````python
# Impute income based on Occupation
df_group = df.copy()
df_group['Income'] = df_group.groupby('Occupation')['Income'].transform(lambda x: x.fillna(x.mean()))

print("\nGroup-by Occupation Imputed:\n")
df_group
````

Group-based imputation (e.g., occupation, age-group) is often preferred in industries because it maintains meaningful differences between groups, making predictions more accurate and realistic

| Median  | Skewed data or outliers present               |

### **Why Handling Outliers is Important in Supervised Learning**

**Outliers** are extreme or unusual data points that differ significantly from the rest.

* **Impact on the model:**
  Outliers can **mislead models**, causing inaccurate predictions. They might skew the model, making it overly influenced by these extreme points.

#### **Industry Example:**

Imagine predicting house prices based on area size.
If there's one unusually priced house (due to luxury features or errors), it could make the model wrongly predict prices for normal homes.

#### **Simple Numeric Example:**

* Data without outlier: `[50, 55, 60, 65, 70]` (Mean = 60)
* Data with outlier: `[50, 55, 60, 65, 250]` (Mean ‚âà 96)

The outlier (250) drastically changes the mean, misleading the model if not handled.

---

**Conclusion:**

> **Handling outliers helps models capture real patterns, leading to more accurate and reliable predictions.**


## **Step-by-Step: Identifying Outliers Using IQR**

### **Step 1: Understand Quartiles**

- **Q1 (First Quartile)**: 25th percentile (25% of data below Q1)
- **Q3 (Third Quartile)**: 75th percentile (75% of data below Q3)
- **IQR (Interquartile Range)**: Range between Q1 and Q3.

$$
\text{IQR} = Q3 - Q1
$$

### **Step 2: Identify Outlier Boundaries**

Outliers are points that fall outside this range:

- **Lower limit:** $Q1 - 1.5 \times IQR$
- **Upper limit:** $Q3 + 1.5 \times IQR$

### **Formula:**

$$
\text{Outlier if } X < Q1 - 1.5 \times IQR \text{ or } X > Q3 + 1.5 \times IQR
$$

---

## **Example with Numeric Data**

Consider the data:

```
10, 12, 13, 14, 15, 16, 18, 50
```

### **Step 1: Calculate Q1 and Q3**

- $Q1 = 12.75$, $Q3 = 16.5$

### **Step 2: Calculate IQR**

$$
IQR = Q3 - Q1 = 16.5 - 12.75 = 3.75
$$

### **Step 3: Set Outlier Limits**

- Lower limit: $12.75 - 1.5 \times 3.75 = 7.125$
- Upper limit: $16.5 + 1.5 \times 3.75 = 22.125$

### **Step 4: Identify Outliers**

- `50` is above `22.125` ‚Üí **Outlier**

---

## **Short: How to Treat Outliers?**

1. **Remove them:**
   If clearly errors or extremely rare events.

2. **Replace (Impute):**
   Use median or capped values to reduce impact.

3. **Transform data:**
   Apply log, square-root transformations to reduce skew.

**Industry recommendation:**
Use domain knowledge to decide whether outliers represent real variations or errors.

# How to Treat Outlier?


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dummy data with outliers
data = np.array([10, 12, 13, 14, 15, 16, 18, 50, 52, 550])
df = pd.DataFrame({'Value': data})

# Calculate Q1, Q3 and IQR
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1

# Set caps (limits)
lower_cap = Q1 - 1.5 * IQR
upper_cap = Q3 + 1.5 * IQR

# Capping (Winsorization)
df['Value_Capped'] = df['Value'].clip(lower=lower_cap, upper=upper_cap)

# Log transformation
df['Value_Log'] = np.log(df['Value'])

# Plotting original vs capped vs log
plt.figure(figsize=(15,4))

# Original data
plt.subplot(1,3,1)
sns.boxplot(y=df['Value'], color='skyblue')
plt.title('Original Data')

# Capped data
plt.subplot(1,3,2)
sns.boxplot(y=df['Value_Capped'], color='lightgreen')
plt.title('Capped Data')

# Log-transformed data
plt.subplot(1,3,3)
sns.boxplot(y=df['Value_Log'], color='lightcoral')
plt.title('Log-transformed Data')

plt.tight_layout()
plt.show()
````

## **1. Normalization & Standardization Formulas**

### **Standardization (Z-score)**:

Transforms data to have mean=0, standard deviation=1.

$$
X_{std} = \frac{X - \mu}{\sigma}
$$

- **X**: original value
- **Œº**: mean of data
- **œÉ**: standard deviation of data

### **Min-Max Normalization**:

Scales data to a specific range, typically \[0,1].

$$
X_{minmax} = \frac{X - X_{min}}{X_{max} - X_{min}}
$$

- **X**: original value
- **X‚Çò·µ¢‚Çô**: minimum value in dataset
- **X‚Çò‚Çê‚Çì**: maximum value in dataset

## **When to Use Label Encoding vs. Ordinal Encoding**

In machine learning, encoding categorical variables into numerical format is essential. Two common methods are **Label Encoding** and **Ordinal Encoding**. However, they have distinct use-cases and implications.

---

## **1. Label Encoding**

### **When to use?**

Use **Label Encoding** when your data has categories **without any inherent order or ranking**.

### **Industry-level Example: Customer Region in Retail**

Consider a retail chain operating in multiple regions:

- North America
- Europe
- Asia
- Australia

These regions **have no natural numeric ordering**. Thus, you can assign arbitrary numeric labels:

| Region        | Encoded Label |
| ------------- | ------------- |
| North America | 0             |
| Europe        | 1             |
| Asia          | 2             |
| Australia     | 3             |

Data scientist can use Label Encoding to transform categorical regions into numeric labels.

---

## **2. Ordinal Encoding**

### **When to use?**

Use **Ordinal Encoding** when categories **have a natural or logical ordering**.

### **Industry-level Example: Credit Rating in Banking**

Banks use credit ratings to assess loan risks. Ratings have a clear, inherent order:

| Credit Rating | Encoded Label |
| ------------- | ------------- |
| Poor          | 0             |
| Fair          | 1             |
| Good          | 2             |
| Excellent     | 3             |

Here, numbers explicitly represent the increasing creditworthiness of a borrower.

### **How a Data Scientist Uses it?**

A data scientist building predictive models, like regression or classification models predicting loan defaults, will use ordinal encoding. This informs the model explicitly that "Excellent" is better than "Good," and "Good" is better than "Fair," etc., ensuring accurate learning.

---

## **Key Differences at a Glance**

| Aspect                        | Label Encoding                    | Ordinal Encoding                                                |
| ----------------------------- | --------------------------------- | --------------------------------------------------------------- |
| **Meaning of Numeric Values** | Arbitrary; no meaningful order    | Meaningful and ordered                                          |
| **Example Use-case**          | Regions, Colors, Brand Names      | Customer Satisfaction Ratings, Credit Ratings, Education Levels |
| **Impact on Models**          | No numeric meaning, no order bias | Explicit order bias, helps numeric models                       |

---

## **Summary (When to Use What)**

- **Label Encoding:**

  - **No meaningful ordering**
  - Examples: Geographic region, Gender, Product categories, Brands
  - Suitable algorithms: Decision Trees, Random Forest, XGBoost

- **Ordinal Encoding:**

  - **Clear meaningful ordering**
  - Examples: Satisfaction level, Risk rating, Priority (Low, Medium, High)
  - Suitable algorithms: Linear Regression, Ordinal Logistic Regression, SVM

---

### **Conclusion:**

Always clearly understand whether your categorical data has a natural order before deciding which encoding method to use. This ensures your model correctly interprets the meaning and relationships between your data points, resulting in better predictions and insights.

## üß† 8.1.3 Model Selection

Model selection is the process of choosing the **best model evaluation strategy** to ensure that our model generalizes well on unseen data. In industry, this is not just about accuracy‚Äî**it's about trust, fairness, stability, and reproducibility.**

Let‚Äôs understand common model selection techniques using **real-world e-commerce and finance use cases**, and end with a comparison table.

---

### ‚úÖ 8.1.3.1 Train‚ÄìTest Split

**What it is:**
Divides the dataset into two parts: a **training set** to build the model and a **testing set** to evaluate performance.

**How it works:**
Typically, 70‚Äì80% of data is used for training and 20‚Äì30% for testing.

**Use Case (E-Commerce):**
When building a **recommendation system**, the data is split to evaluate how well the system performs on new users.

**Limitation:**
A single split might result in **biased or unstable results**, especially with small datasets.

---

### ‚úÖ 8.1.3.2 Cross-Validation (Cross Val Score)

**What it is:**
A method where the dataset is split into multiple folds (usually 5 or 10), and the model is trained and validated **k times**, each time using a different fold as the validation set.

**How it works:**
The final score is the **average** of all validation scores.

**Use Case (Finance):**
Predicting **loan default risk** ‚Äî ensures the model isn't biased toward a specific sample and performs well across data segments.

**Benefit:**
Reduces variance due to train-test splits; helps choose between multiple models.

---

### ‚úÖ 8.1.3.3 K-Fold Cross-Validation

**What it is:**
A specific form of cross-validation where the data is divided into **K equal-sized folds**.

**How it works:**
Each fold is used as a validation set **once**, and the remaining **K-1** folds form the training set.

**Use Case (Healthcare):**
Building a **disease prediction model** from patient data. K-Fold ensures all records are used for both training and validation.

**Best For:**
Large, balanced datasets with **independent observations**.

---

### ‚úÖ 8.1.3.4 Repeated K-Fold

**What it is:**
Repeats K-Fold cross-validation **multiple times** with different random splits, giving a **more reliable estimate** of performance.

**How it works:**
You choose `K` and how many times to repeat it (e.g., 10 folds repeated 3 times).

**Use Case (Retail):**
Forecasting **weekly sales trends** from past data where results may vary by sampling. Repetition reduces randomness.

**Advantage:**
Improves **model robustness** by averaging results from many splits.

---

### ‚úÖ 8.1.3.5 Stratified K-Fold

**What it is:**
Variation of K-Fold that **preserves the class distribution** in each fold‚Äîimportant for **classification tasks**.

**How it works:**
Ensures that the proportion of each class (e.g., churned vs. active customers) is roughly the same across all folds.

**Use Case (Customer Churn):**
Churn datasets are often imbalanced. Stratified K-Fold ensures fair training and evaluation.

**Why Use:**
Prevents models from **overfitting the majority class**.

---

### ‚úÖ 8.1.3.6 Repeated Stratified K-Fold

**What it is:**
Combines the **class-balancing benefits** of Stratified K-Fold with the **stability of repetition**.

**Use Case (Insurance):**
Predicting rare events like **fraud detection**, where class imbalance is high. Repeated stratification ensures both balance and multiple evaluations.

**Best For:**
Small or imbalanced classification datasets where **accuracy, precision, and recall** matter.

---

### üßæ Comparison Table

| Technique                  | Balances Data? | Repeats Splits? | Best Use Case                        | Limitation                     |
| -------------------------- | -------------- | --------------- | ------------------------------------ | ------------------------------ |
| Train‚ÄìTest Split           | ‚ùå             | ‚ùå              | Quick experiments, large datasets    | High variance                  |
| Cross Val Score            | ‚ùå             | ‚úÖ              | Model selection, general cases       | May not balance classes        |
| K-Fold                     | ‚ùå             | ‚ùå              | Generic regression/classification    | Unbalanced folds possible      |
| Repeated K-Fold            | ‚ùå             | ‚úÖ              | Time-series & repeated trials        | May still ignore class balance |
| Stratified K-Fold          | ‚úÖ             | ‚ùå              | Imbalanced classification problems   | Still single evaluation cycle  |
| Repeated Stratified K-Fold | ‚úÖ             | ‚úÖ              | Fraud, churn, rare disease detection | Computationally expensive      |


## üöÄ 8.1.4 Model Deployment: Linear Regression (Manual Calculation Demo)

**Linear Regression** is one of the simplest and most widely used algorithms in machine learning. It models the relationship between two variables by fitting a **straight line** to the data.

### üìå Objective:

Fit a straight line:

$$
y = mx + c
$$

Where:

- $m$ is the slope (how much y changes with x)
- $c$ is the intercept (y-value when x = 0)

---

### üõí Dataset: Mall Customers (6 Rows Sample)

Let‚Äôs use a tiny dataset of customers where we want to predict **Annual Spend (‚Çπ)** based on **Age**.

| Customer | Age (x) | Annual Spend (‚Çπy) |
| -------- | ------- | ----------------- |
| C1       | 25      | 40,000            |
| C2       | 30      | 50,000            |
| C3       | 35      | 55,000            |
| C4       | 40      | 65,000            |
| C5       | 45      | 70,000            |
| C6       | 50      | 75,000            |

---

### üßÆ Step-by-Step Calculation:

We will compute the following to find slope (m) and intercept (c):

| x (Age) | y (Spend) |   x¬≤ |        xy |
| ------: | --------: | ---: | --------: |
|      25 |     40000 |  625 | 1,000,000 |
|      30 |     50000 |  900 | 1,500,000 |
|      35 |     55000 | 1225 | 1,925,000 |
|      40 |     65000 | 1600 | 2,600,000 |
|      45 |     70000 | 2025 | 3,150,000 |
|      50 |     75000 | 2500 | 3,750,000 |

#### ‚úÖ Summations:

$$
\sum x = 225,\quad \sum y = 355000,\quad \sum x^2 = 8875,\quad \sum xy = 13,925,000,\quad n = 6
$$

---

### üìè Step 1: Calculate Slope (m)

$$
m = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2}
$$

Substitute values:

$$
m = \frac{6 \cdot 13925000 - 225 \cdot 355000}{6 \cdot 8875 - 225^2}
$$

$$
m = \frac{83550000 - 79875000}{53250 - 50625} = \frac{3675000}{2625} = 1400
$$

---

### üìç Step 2: Calculate Intercept (c)

$$
c = \frac{\sum y - m \sum x}{n} = \frac{355000 - 1400 \cdot 225}{6}
$$

$$
c = \frac{355000 - 315000}{6} = \frac{40000}{6} ‚âà 6666.67
$$

---

### ‚úÖ Final Linear Equation:

$$
\boxed{y = 1400x + 6666.67}
$$

This means:
**For every 1-year increase in age, the predicted annual spend increases by ‚Çπ1400.**

---

### üîç Model Deployment Use Case (Retail Marketing)

In an industry setting:

- **Input:** Age of a new customer.
- **Model:** $y = 1400x + 6666.67$
- **Output:** Predicted spending.

**Example:**
If a new customer is 38 years old, expected annual spend:

$$
y = 1400 \cdot 38 + 6666.67 = 53200 + 6666.67 ‚âà ‚Çπ59866.67
$$


````python


## üéØ Real-World Example: Predicting Sales Based on Advertising Spend


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import skew

# Set seed for reproducibility
np.random.seed(42)

# Create advertising budget (in ‚Çπ1000s)
ad_spend = np.random.randint(10, 100, size=20)

# Sales = 2.5 * ad_spend + noise
noise = np.random.normal(0, 10, size=20)
sales = 2.5 * ad_spend + noise

# Create DataFrame
df = pd.DataFrame({'Ad_Spend': ad_spend, 'Sales': sales})
df.head()


````

````python
plt.figure(figsize=(6, 4))
plt.scatter(df['Ad_Spend'], df['Sales'], color='blue')
plt.title('Ad Spend vs Sales')
plt.xlabel('Ad Spend (‚Çπ1000s)')
plt.ylabel('Sales (‚Çπ1000s)')
plt.grid(True)
plt.show()
````

````python

### ü§ñ 3. Train Linear Regression Model


# Reshape input
X = df[['Ad_Spend']]
y = df['Sales']

model = LinearRegression()
model.fit(X, y)

# Predictions
df['Predicted_Sales'] = model.predict(X)

df
````

````python

### üìê 4. Plot the Regression Line


plt.figure(figsize=(6, 4))
plt.scatter(df['Ad_Spend'], df['Sales'], label='Actual', color='blue')
plt.plot(df['Ad_Spend'], df['Predicted_Sales'], label='Regression Line', color='red')
plt.xlabel('Ad Spend (‚Çπ1000s)')
plt.ylabel('Sales (‚Çπ1000s)')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
````

````python

### üìä 5. Evaluate the Model

mse = mean_squared_error(y, df['Predicted_Sales'])
r2 = r2_score(y, df['Predicted_Sales'])

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R¬≤ Score: {r2:.2f}")
````

- **R¬≤ (Coefficient of Determination)**: Proportion of variance in Sales explained by Ad Spend. Closer to 1 is better.

### üìå **Mean Absolute Error (MAE)**

**‚úÖ Formula:**

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|
$$

**üéØ Range:**

$$
[0, \infty)
$$

**üß† When to Use:**

- When you want to treat **all errors equally**.
- When you want an **easy-to-understand** metric (same units as the target).
- Less sensitive to outliers than MSE or RMSE.

**‚úÖ Examples:**

- Forecasting demand in retail.
- Predicting housing or product prices without extreme fluctuations.

---

### üìå **Median Absolute Error (MedAE)**

**‚úÖ Formula:**

$$
\text{MedAE} = \text{median} \left( \left| y_i - \hat{y}_i \right| \right)
$$

**üéØ Range:**

$$
[0, \infty)
$$

**üß† When to Use:**

- When your data contains **outliers** or **non-normal error distributions**.
- When you want a **robust measure** not affected by extreme errors.

**‚úÖ Examples:**

- Predicting healthcare costs (where rare high-cost patients exist).
- Real estate models including luxury properties that skew prices.


### üìå **1. Mean Absolute Error (MAE)**

**‚úÖ Formula:**

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|
$$

**üéØ Range:**

$$
[0, \infty)
$$

**üß† When to Use:**

- When all errors should be **weighted equally**.
- For **easier interpretation** (same units as target).
- Less affected by **outliers** compared to squared metrics.

**üíº Use Cases:**

- Retail sales forecasting
- Budget planning
- Predicting delivery times

---

### üìå **2. Median Absolute Error (MedAE)**

**‚úÖ Formula:**

$$
\text{MedAE} = \text{median} \left( \left| y_i - \hat{y}_i \right| \right)
$$

**üéØ Range:**

$$
[0, \infty)
$$

**üß† When to Use:**

- Data contains **extreme outliers**.
- You want a **robust metric** not skewed by rare events.

**üíº Use Cases:**

- Healthcare cost prediction
- Real estate pricing (with rare high-end properties)
- Legal settlement estimations

---

### üìå **3. Mean Squared Error (MSE)**

**‚úÖ Formula:**

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

**üéØ Range:**

$$
[0, \infty)
$$

**üß† When to Use:**

- When **larger errors must be penalized more**.
- Suitable for **model optimization** (differentiable).
- Sensitive to **outliers**, so use when you care about them.

**üíº Use Cases:**

- Credit risk modeling
- Time-series forecasting
- Predicting rare equipment failures

---

### üìå **4. Root Mean Squared Error (RMSE)**

**‚úÖ Formula:**

$$
\text{RMSE} = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 }
$$

**üéØ Range:**

$$
[0, \infty)
$$

**üß† When to Use:**

- When you need **error in original units** (e.g., ‚Çπ, ¬∞C).
- Same use cases as MSE but **more interpretable**.
- Still penalizes large errors heavily.

**üíº Use Cases:**

- Weather forecasting (temperature)
- Energy consumption predictions
- Stock price prediction

---

### üìå **5. R¬≤ Score (Coefficient of Determination)**

**‚úÖ Formula:**

$$
R^2 = 1 - \frac{ \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 }{ \sum_{i=1}^{n} (y_i - \bar{y})^2 }
$$

**üéØ Range:**

$$
(-\infty, 1]
$$

**üß† When to Use:**

- To evaluate **how much variance** the model explains.
- Good for **comparing models**.
- Should **not be used alone** ‚Äî pair with MAE, RMSE, etc.

**üíº Use Cases:**

- Financial performance prediction
- Marketing ROI modeling
- User engagement prediction (CTR, retention)


#** Assumptions of Linear Regression**

When we use **Linear Regression** (a method to predict numerical outcomes), we make certain assumptions. These assumptions are like the "rules" that should ideally hold true for the model to give accurate predictions.

---

## **1. Linearity**

- **Meaning:**
  The relationship between input features (independent variables) and output (dependent variable) should be approximately **linear** (a straight line relationship).

- **Industry Example (Real Estate):**
  When predicting house prices, we assume a clear, direct relationship between size (sq ft) and price.
  **More size ‚Üí higher price (in a straight-line manner).**

---

## **2. Independence of Observations**

- **Meaning:**
  Each data point (observation) should be independent of others. Observing one value shouldn't influence another.

- **Industry Example (Marketing Campaigns):**
  Customer responses to a survey should be independent. One customer's opinion shouldn‚Äôt directly affect another's.

---

## **3. Homoscedasticity (Constant Variance)**

- **Meaning:**
  Errors (difference between actual and predicted values) should have similar variance across all levels of input values.

- **Industry Example (Sales Forecasting):**
  If predicting monthly sales based on advertising spend, the errors (prediction misses) should stay relatively consistent at all spending levels‚Äînot much higher errors for higher spend.

---

## **4. Normality of Residuals**

- **Meaning:**
  Residuals (errors between actual and predicted values) should follow roughly a normal distribution (bell-shaped curve).

- **Industry Example (Healthcare):**
  Predicting patient recovery time based on age. Errors (predicted vs actual recovery times) ideally cluster around zero, forming a symmetrical bell shape.

  If not followed:
  Confidence intervals and hypothesis tests may not be accurate, affecting conclusions and model interpretations.

---

## **5. No Multicollinearity**

- **Meaning:**
  Independent variables shouldn‚Äôt be highly correlated with each other. Each feature should give unique information.

- **Industry Example (Financial Industry):**
  Predicting stock prices using indicators like interest rates and inflation. If interest rate and inflation are highly related, it becomes difficult for the model to clearly distinguish their individual effects.

---

| Assumption               | Industry Example       | If assumption fails‚Ä¶                     |
| ------------------------ | ---------------------- | ---------------------------------------- |
| **Linearity**            | House price vs. size   | Poor predictions; biased estimates       |
| **Independence**         | Customer surveys       | Misleading significance tests            |
| **Homoscedasticity**     | Advertising vs. sales  | Unreliable predictions at certain ranges |
| **Normality**            | Patient recovery times | Incorrect confidence intervals/tests     |
| **No Multicollinearity** | Economic indicators    | Instable coefficient estimates           |

---

### **Why are these assumptions important?**

They help ensure your linear regression model accurately represents real-world relationships, producing reliable, useful predictions.
