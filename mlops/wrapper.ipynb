```python
import functools
def tracktime(func):
    @functools.wraps(func)
    def track(*args , **kwargs):
        print(f'Starting {func.__name__}')
        start = time.time()
        result = func(*args , **kwargs)
        end = time.time()
        diff = round(end - start , 2)
        print(f'{result['pro']} completed in {diff} seconds')

    return track
```

```python
@tracktime
def load_data():
    '''Loads data to the model'''
    time.sleep(2)
    return {'pro': 'load_data'}

@tracktime
def train_model():
    '''Trains the model'''
    time.sleep(5)
    return {'pro':'train_model'}


@tracktime
def eval_model():
    '''Evaluate the model'''
    time.sleep(5)
    return {'pro':'eval_model'}
```
