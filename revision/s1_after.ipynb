# Linear Regression


```python
w = np.zeros((1 , 5))
x = np.random.random((10 , 5))
(x @ w.T).shape
```

```python
def lr(x , y , alpha , niter):
    #make sure y is a column vector:
    y = y.reshape(-1 , 1)
    n , m = x.shape
    theta = np.zeros((m , 1))

    for iter in range(niter):
        ypred = x @ theta
        
        if iter == 0: 
            print(ypred.shape , y.shape)
            print((ypred - y).shape)
        dw =  -(x.T @ (ypred - y)) / n
        if iter == 0 : print(dw.shape)
        theta += alpha * dw

    return np.round(theta , 4)

x = np.array([[1, 1], [1, 2], [1, 3]])
y = np.array([1, 2, 3])
alpha = 0.01
iterations = 1000

lr(x , y , alpha , iterations)
```
