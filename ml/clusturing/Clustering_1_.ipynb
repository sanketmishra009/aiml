# üîç Introduction to Clustering in the Real World

In today‚Äôs data-driven world, companies are flooded with massive amounts of information‚Äîmillions of customer transactions, sensor readings, web activity logs, and more. But raw data alone is not useful until we find patterns or structure in it.

That‚Äôs where **clustering** comes in.

Clustering is the process of grouping similar data points together. These groups are called **clusters**, and each group contains data points that are more alike to each other than to points in other groups. Unlike classification, clustering works with **unlabelled data**‚Äîwe don‚Äôt tell the algorithm what to look for; we let the data speak for itself.

---

## üì¶ Why Clustering Matters in Industry

Clustering is widely used in the industry for:

| Industry        | Use Case                                                                               |
| --------------- | -------------------------------------------------------------------------------------- |
| üõçÔ∏è Retail       | Segmenting customers based on shopping behavior to create targeted marketing campaigns |
| üí≥ Banking      | Detecting unusual customer behavior to flag potential fraud                            |
| üöó Automotive   | Grouping sensor data to identify patterns that may lead to predictive maintenance      |
| üß¨ Healthcare   | Grouping patients based on medical records for personalized treatment plans            |
| üì± Social Media | Grouping users based on their activity, preferences, and interactions                  |

---

# üß† What is k-Means Clustering?

k-Means is one of the most popular clustering algorithms used in industry.

It works like this:

1. **Choose `k`** (the number of clusters you want).
2. **Randomly assign** `k` points as cluster centers (called **centroids**).
3. **Assign each data point** to the cluster whose centroid is nearest.
4. **Update** the centroids by calculating the average of the points in each cluster.
5. **Repeat** steps 3 and 4 until the clusters stop changing (or reach a max number of iterations).

---

## üìå Real-World Example: Telecom Industry

Suppose you are a data analyst at a telecom company. You have millions of customer records with data like:

- Monthly data usage
- Average call duration
- Number of complaints
- Bill payment behavior

You want to divide customers into distinct groups so that:

- **High-value customers** can be offered loyalty benefits
- **At-risk customers** can be contacted before they churn
- **Late payers** can be offered auto-pay discounts

k-Means can help you **group these customers** based on their similarity‚Äîno need for manual tagging.

---

# üìê How Similarity is Measured

The **distance** between two data points determines how similar they are. A common distance measure is **Euclidean Distance**, which you can think of as the straight-line distance between two points.

If customers A and B have very similar data usage and bill amounts, the distance between them is small‚Äîthey‚Äôre likely to end up in the same cluster.

> üìå The shorter the distance, the **more similar** the data points.

---

# üîÑ What is a Centroid?

A **centroid** is like the "center of gravity" of a cluster. It‚Äôs the average of all the points in a cluster and acts as the **representative** of that group.

Each time points move between clusters during the algorithm's steps, the centroids are updated to reflect the new average positions.

---

# üéØ Goal of k-Means

The algorithm tries to:

- **Minimize Intra-cluster distance**: Keep data points in the same cluster as close to the centroid as possible.
- **Maximize Inter-cluster distance**: Ensure that clusters are as far apart from each other as possible.

---

## üìä Key Metrics in k-Means

To evaluate how good your clusters are, we use these metrics:

### ‚úÖ WCSS ‚Äì Within-Cluster Sum of Squares

- Measures how close the data points in a cluster are to the centroid.
- Lower WCSS = tighter, more compact clusters.
- But: WCSS **increases with more data points**, so not ideal for comparing clusters of different sizes.

### ‚úÖ BSS ‚Äì Between-Cluster Sum of Squares

- Measures how far apart the centroids of different clusters are from the overall average.
- Higher BSS = better-separated clusters.

### ‚úÖ TSS ‚Äì Total Sum of Squares

- TSS = WCSS + BSS.
- Represents total variability in the dataset.

---

# ‚ùì How to Choose the Right Number of Clusters (`k`)?

Choosing the wrong `k` can ruin your results:

- Too few clusters: You lose meaningful separation.
- Too many clusters: You end up with noise and overfitting.

### üìâ Elbow Method

This is a visual way to find the right `k`. You:

1. Run k-means for different values of `k`.
2. Plot `k` vs. WCSS.
3. Look for the ‚Äúelbow‚Äù in the curve‚Äîwhere WCSS starts decreasing slowly. That‚Äôs your ideal `k`.

---

# Cluster Analysis: Step-by-Step Example with Calculations

Suppose we have customer spending data (e.g., average monthly purchase in ‚Çπ1000s):

```
[2, 3, 4, 10, 11, 12, 20, 25, 30]
```

We aim to create **2 clusters** using different combinations and compute:

- **Within-Cluster Sum of Squares (WCSS)**
- **Between-Cluster Sum of Squares (BCSS)**

We will compare 3 clustering scenarios and select the best combination based on the following criteria:

- **Low WCSS** (compact clusters)
- **High BCSS** (well-separated clusters)

---

## Step 1: Compute the Overall Mean (Sample Mean)

To calculate **BCSS**, we first compute the overall mean (grand mean):

$$
\text{Sample Mean} = \frac{\sum x_i}{n} = \frac{2 + 3 + 4 + 10 + 11 + 12 + 20 + 25 + 30}{9} = \frac{117}{9} = 13
$$

---

## Case 1: Clusters = \[2, 3, 4, 10, 11] and \[12, 20, 25, 30]

### Cluster 1:

- Elements: \[2, 3, 4, 10, 11]
- Centroid: $\frac{2 + 3 + 4 + 10 + 11}{5} = 6$

### Cluster 2:

- Elements: \[12, 20, 25, 30]
- Centroid: $\frac{12 + 20 + 25 + 30}{4} = 21.75$

### WCSS Calculation

#### Cluster 1:

$$
(2 - 6)^2 + (3 - 6)^2 + (4 - 6)^2 + (10 - 6)^2 + (11 - 6)^2 \\
= 16 + 9 + 4 + 16 + 25 = 70
$$

#### Cluster 2:

$$
(12 - 21.75)^2 + (20 - 21.75)^2 + (25 - 21.75)^2 + (30 - 21.75)^2 \\
= 95.06 + 3.06 + 10.56 + 68.06 = 176.75
$$

**Total WCSS (Case 1)** = 70 + 176.75 = **246.75**

### BCSS Calculation

$$
\text{BCSS} = 5(6 - 13)^2 + 4(21.75 - 13)^2 = 5(49) + 4(76.5625) = 245 + 306.25 = 551.25
$$

---

## Case 2: Clusters = \[2, 3, 4, 10] and \[11, 12, 20, 25, 30]

### Cluster 1:

- Elements: \[2, 3, 4, 10]
- Centroid: $\frac{2 + 3 + 4 + 10}{4} = 4.75$

### Cluster 2:

- Elements: \[11, 12, 20, 25, 30]
- Centroid: $\frac{11 + 12 + 20 + 25 + 30}{5} = 19.6$

### WCSS Calculation

#### Cluster 1:

$$
(2 - 4.75)^2 + (3 - 4.75)^2 + (4 - 4.75)^2 + (10 - 4.75)^2 \\
= 7.5625 + 3.0625 + 0.5625 + 27.5625 = 38.75
$$

#### Cluster 2:

$$
(11 - 19.6)^2 + (12 - 19.6)^2 + (20 - 19.6)^2 + (25 - 19.6)^2 + (30 - 19.6)^2 \\
= 73.96 + 57.76 + 0.16 + 29.16 + 108.16 = 269.2
$$

**Total WCSS (Case 2)** = 38.75 + 269.2 = **307.95**

### BCSS Calculation

$$
\text{BCSS} = 4(4.75 - 13)^2 + 5(19.6 - 13)^2 = 4(68.0625) + 5(43.56) = 272.25 + 217.8 = 490.05
$$

---

## Case 3: Clusters = \[2, 3, 4, 10, 11, 12] and \[20, 25, 30]

### Cluster 1:

- Elements: \[2, 3, 4, 10, 11, 12]
- Centroid: $\frac{2 + 3 + 4 + 10 + 11 + 12}{6} = 7$

### Cluster 2:

- Elements: \[20, 25, 30]
- Centroid: $\frac{20 + 25 + 30}{3} = 25$

### WCSS Calculation

#### Cluster 1:

$$
(2 - 7)^2 + (3 - 7)^2 + (4 - 7)^2 + (10 - 7)^2 + (11 - 7)^2 + (12 - 7)^2 \\
= 25 + 16 + 9 + 9 + 16 + 25 = 100
$$

#### Cluster 2:

$$
(20 - 25)^2 + (25 - 25)^2 + (30 - 25)^2 \\
= 25 + 0 + 25 = 50
$$

**Total WCSS (Case 3)** = 100 + 50 = **150**

### BCSS Calculation

$$
\text{BCSS} = 6(7 - 13)^2 + 3(25 - 13)^2 = 6(36) + 3(144) = 216 + 432 = 648
$$

---

## Summary Table:

| Case | Cluster Sizes | Centroids   | WCSS   | BCSS   |
| ---- | ------------- | ----------- | ------ | ------ |
| 1    | 5 / 4         | 6 / 21.75   | 246.75 | 551.25 |
| 2    | 4 / 5         | 4.75 / 19.6 | 307.95 | 490.05 |
| 3    | 6 / 3         | 7 / 25      | 150    | 648    |

---

## Final Conclusion

Based on the comparison:

- **Case 3** has the **lowest WCSS (150)** ‚Üí clusters are compact
- **Case 3** has the **highest BCSS (648)** ‚Üí clusters are well separated

# Hierarchical Clustering ‚Äì


Hierarchical Clustering ‚Äì

Hierarchical clustering is a **clustering technique** used to group similar data points together into **clusters**. It is called _‚Äúhierarchical‚Äù_ because it builds a hierarchy (a tree-like structure) of clusters ‚Äî either by **merging smaller clusters into bigger ones** (bottom-up) or **dividing a big cluster into smaller ones** (top-down).

---

## üî∂ Real-World Example: Customer Segmentation for a Retail Company

Imagine you're working as a data analyst for a retail company like **Reliance Trends** or **Pantaloons**. You have a **small sample of customer data (say 15-30 customers)** with features like:

- Age
- Annual spending
- Frequency of purchases

Your goal: **Group similar customers into clusters** so marketing teams can create personalized campaigns.

Because the sample size is small, we use **Hierarchical Clustering** (not K-Means, which works better for large datasets).

---

## üî∑ Two Main Types of Hierarchical Clustering

### 1. **Agglomerative (Bottom-Up)** ‚Äì Most Common

- Each data point starts in its own cluster.
- At each step, the two **most similar** clusters are merged.
- Repeats until there‚Äôs **one single cluster** (all data points merged).
- Example: You start with 15 customers, and keep merging until you have 1 mega group.

üü© This is great when you want to **see all possible clustering levels** (from 15 clusters to 1).

### 2. **Divisive (Top-Down)** ‚Äì Less Common

- Start with all data points in **one big cluster**.
- Split the most dissimilar ones into two groups.
- Continue splitting each group until each point is its own cluster.

---

## üî∑ Steps in Agglomerative Hierarchical Clustering

Let‚Äôs break down how Agglomerative Hierarchical Clustering is performed step-by-step:

---

### üîπ Step 1: Normalize the Data

All features (like Age, Spending, etc.) must be on the **same scale**. For example, Age (18‚Äì65) and Spending (‚Çπ10,000‚Äì‚Çπ1,00,000) are different in range ‚Äî this could mislead the distance calculation.

üìå We use **MinMaxScaler** or **StandardScaler** to normalize data.

---

### üîπ Step 2: Calculate the Distance Between Points

Use **Euclidean Distance** (most common) or any distance measure.

### üßÆ Example:

Let‚Äôs say we have two customers:

| Customer | Age | Spending |
| -------- | --- | -------- |
| A        | 25  | ‚Çπ40,000  |
| B        | 30  | ‚Çπ42,000  |

Euclidean Distance = ‚àö\[(30‚Äì25)¬≤ + (42000‚Äì40000)¬≤]
\= ‚àö\[25 + 4,000,000] ‚âà 2000.006 units
(Since one unit of spending = ‚Çπ1, Euclidean distance reflects actual difference.)

---

### üîπ Step 3: Choose Agglomeration Method

This step determines how clusters are merged. There are **four popular methods**:

| Method              | How It Works                                                                    |
| ------------------- | ------------------------------------------------------------------------------- |
| **Single-Link**     | Min distance between any 2 points (one from each cluster)                       |
| **Complete-Link**   | Max distance between any 2 points (one from each cluster)                       |
| **Average-Link**    | Average of all pairwise distances between two clusters                          |
| **Centroid Method** | Use the centroid (mean point) of each cluster and compute distance between them |

üìå **Average-link and complete-link** are often used in customer segmentation.

---

### üîπ Step 4: Pass Distance + Method into Clustering Algorithm

Once distance and linkage type are defined, the algorithm keeps **merging clusters step-by-step** using your selected logic.

---

### üîπ Step 5: Plot the Dendrogram üìä

A **dendrogram** is a tree diagram that shows how clusters are merged.

- **X-axis:** Customer IDs or points
- **Y-axis:** Distance between clusters (i.e., similarity)

The **higher the horizontal line**, the **less similar** the clusters.

---

### üîª Example: Visualizing Customer Clustering Using Dendrogram

```
             |
             |
             |         _________
             |        |         |
             |    ____|____     |
             |   |         |    |
             |___|___   ___|____|__
                 A B   C     D   E
```

Here:

- Customers A and B are **very similar** (joined at a low height).
- C joins them next.
- D and E are more dissimilar (join at a higher height).
- The longer the line between clusters, the **greater the difference**.

üìå You can **draw a horizontal line** (cut) across the dendrogram to get **optimal number of clusters**.

---

## ‚úÖ Final Output Example: Marketing Segmentation

Let‚Äôs say the dendrogram gives you 3 clusters of customers:

| Cluster | Customer Behavior                      |
| ------- | -------------------------------------- |
| A       | Young, High Spend, Frequent Buyers     |
| B       | Middle-aged, Average Spend, Occasional |
| C       | Older, Low Spend, Infrequent           |

The marketing team can now:

- Offer **premium loyalty programs** to Cluster A
- Send **reminders or offers** to Cluster B
- Run **retargeting campaigns** for Cluster C

---

## üß† Summary Table: Hierarchical Clustering

| Feature        | Agglomerative                 | Divisive                  |
| -------------- | ----------------------------- | ------------------------- |
| Approach       | Bottom-up                     | Top-down                  |
| Starting point | Each point is its own cluster | All points in one cluster |
| Common usage   | ‚úÖ Yes                        | ‚ùå Rare                   |
| Dendrogram     | Yes (used for both)           | Yes                       |
| Suitable for   | Small datasets                | Large, but rarely used    |
| Key metric     | Distance + Linkage method     | Distance + Splitting rule |

## ‚úÖ When to Use Hierarchical Clustering

- When dataset is small (< 100‚Äì200 points)
- You need **detailed cluster hierarchy**
- You want to **visualize clusters and make decisions manually**


# K Means on Real DataSet


## üöó K-Means Clustering on Driver Behavior Data ‚Äì Introduction

### üî∂ Problem Statement

A transport or logistics company (e.g., **Uber Freight**, **Delhivery**, or **RedBus**) wants to **segment its drivers** based on their driving behavior to:

- **Reward safe drivers**
- **Identify risky drivers**
- **Optimize training and resource allocation**

They collect **daily telematics data** such as:

- Average distance driven
- Overspeeding frequency

To gain actionable insights, we‚Äôll use **K-Means Clustering**, a popular **unsupervised machine learning algorithm**, to group drivers based on their **similarity in driving patterns**.

---

## üìä Understanding the Dataset

We are using the following dataset:

üìÅ GitHub URL:
`https://raw.githubusercontent.com/JangirSumit/kmeans-clustering/refs/heads/master/driver-data.csv`

Let‚Äôs first understand what each column means:

| Column Name            | Description                                                       |
| ---------------------- | ----------------------------------------------------------------- |
| `id`                   | Unique identifier for each driver                                 |
| `mean_dist_day`        | Average distance driven per day by the driver (in km)             |
| `mean_over_speed_perc` | Percentage of time the driver was over-speeding on average (in %) |

üß† **Goal**: Identify natural clusters of drivers ‚Äî such as:

- üöô Efficient + safe drivers
- üöÄ Fast but risky drivers
- üê¢ Low-distance slow drivers


````python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/JangirSumit/kmeans-clustering/refs/heads/master/driver-data.csv"
df = pd.read_csv(url)

# Preview data
print(df.head())
print(df.describe())
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='mean_dist_day', y='mean_over_speed_perc', s=60, color='blue', edgecolor='black')
plt.title('Driver Behavior Scatter Plot')
plt.xlabel('Mean Distance per Day (km)')
plt.ylabel('Mean Over-Speeding (%)')
plt.grid(True)
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score


# Select features for clustering
X = df[['mean_dist_day', 'mean_over_speed_perc']]

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow Method to determine the optimal number of clusters
wcss = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot WCSS (Elbow Method)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(K_range, wcss, marker='o')
plt.title('Elbow Method - Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.grid(True)

# Plot Silhouette Scores
plt.subplot(1, 2, 2)
plt.plot(K_range, silhouette_scores, marker='o', color='green')
plt.title('Silhouette Score Analysis')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Silhouette Score')
plt.grid(True)

plt.tight_layout()
plt.show()

# Return best K based on max silhouette score
optimal_k = K_range[silhouette_scores.index(max(silhouette_scores))]
optimal_k

````

### üìä K-Means Clustering Analysis

#### ‚úÖ Step 1: **Elbow Method (WCSS - Within-Cluster Sum of Squares)**

- We calculated WCSS for different values of K (2 to 10).
- WCSS tends to decrease with increasing K because adding more clusters reduces the distance within each cluster.
- The goal is to find the ‚Äúelbow‚Äù point ‚Äî where the rate of decrease sharply slows down, indicating that adding more clusters doesn‚Äôt significantly improve compactness.

**Observation:**
In the plot, the elbow appears around **K = 4**, suggesting that 4 clusters balance compactness and simplicity.

---

#### ‚úÖ Step 2: **Silhouette Score Analysis**

- The **Silhouette Score** measures how similar a data point is to its own cluster (cohesion) compared to other clusters (separation).
- It ranges from -1 to 1:

  - **Close to 1** ‚Üí well-clustered.
  - **Close to 0** ‚Üí overlapping clusters.
  - **Negative** ‚Üí wrong cluster assignment.

**Result:**

- The highest Silhouette Score occurs at **K = 4**, confirming that it is the best choice.

Here's a detailed analysis of Hierarchical Clustering applied using four linkage methods‚Äî**single**, **complete**, **average**, and **centroid**‚Äîon the driver dataset (`mean_dist_day`, `mean_over_speed_perc`). Below is an explanation of what each linkage method does, how to interpret its dendrogram, and which method might be most meaningful.

---

### üîç **Linkage Methods Explained**

| Linkage Type         | Description                                                  | Behavior                                                               |
| -------------------- | ------------------------------------------------------------ | ---------------------------------------------------------------------- |
| **Single Linkage**   | Distance between the closest pair of points in two clusters  | Tends to form long, "chain-like" clusters; sensitive to noise/outliers |
| **Complete Linkage** | Distance between the farthest pair of points in two clusters | Produces compact, spherical clusters; avoids chaining                  |
| **Average Linkage**  | Average distance between all pairs of points in two clusters | Balances between single and complete linkage                           |
| **Centroid Linkage** | Distance between centroids of two clusters                   | Can cause inversions in the dendrogram (non-monotonic)                 |

---

### üìä **Dendrogram Analysis Guide**

1. **Cut the dendrogram** at a certain height (distance) to determine the number of clusters.
2. A **long vertical line before a merge** indicates well-separated clusters.
3. **Horizontal cuts** below that line determine your final clusters.

---

### üìà **Observations from Dendrograms**

#### 1. **Single Linkage**

- **Shape:** Shows many short merges and one or two long vertical lines.
- **Interpretation:** Forms one big cluster gradually; chaining effect.
- **Conclusion:** Not ideal in this case because the clusters seem to be elongated rather than compact.

#### 2. **Complete Linkage**

- **Shape:** Clear, long vertical lines separating clusters.
- **Interpretation:** More compact and distinct clusters.
- **Conclusion:** Best for identifying well-separated compact groups; likely a good choice here.

#### 3. **Average Linkage**

- **Shape:** Balance between single and complete.
- **Interpretation:** Good overall; doesn't favor elongated or compact clusters.
- **Conclusion:** Second-best option; works well when data isn't strongly spherical or chained.

#### 4. **Centroid Linkage**

- **Shape:** Some non-monotonic steps (zigzag-like merges).
- **Interpretation:** May show inversions; clusters sometimes appear inconsistent.
- **Conclusion:** Not ideal; less stable and interpretable than average or complete linkage.

---

### ‚úÖ **Which Linkage Makes More Sense?**

- **Best:** `Complete Linkage` ‚Äî offers the clearest and most meaningful cluster separation.
- **Runner-up:** `Average Linkage` ‚Äî balances extremes, handles variance well.
- **Avoid:** `Single` (due to chaining) and `Centroid` (due to possible inversions).

---

### üß† **How to Use This in Practice**

- Always **standardize** your data before hierarchical clustering.
- Try multiple linkages and visually inspect dendrograms.
- Select linkage with **clear vertical separations** and **minimal chaining or inversions**.
- Use dendrogram "cutting" to choose the number of clusters‚Äîtypically where the vertical jump is highest.

````python


from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from collections import Counter
# Load the dataset
df = pd.read_csv("train.csv", header=None, names=["Item ID", "Sentiment", "Tweet"],encoding='latin-1')

# Extract tweet text
tweet_list = df["Tweet"].tolist()

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Determine sentiment for each tweet
tweetsWithSent = []
pos_num = 0
neg_num = 0
neu_num = 0

for text in tweet_list:
    ps = analyzer.polarity_scores(text)
    tweetsWithSent.append({'text': text, 'compound': ps['compound']})

# Create DataFrame of tweets and their polarity
tweet_df = pd.DataFrame(tweetsWithSent)

# Determine sentiment label based on compound score
sentiment = []
for i in tweet_df["compound"]:
    if i > 0:
        sentiment.append("Positive")
        pos_num += 1
    elif i < 0:
        sentiment.append("Negative")
        neg_num += 1
    else:
        sentiment.append("Neutral")
        neu_num += 1

# Add sentiment column to DataFrame
tweet_df["sentiment"] = sentiment

# Display results
print("Displaying first few records:")
print(tweet_df.head())
print("\nSentiment Distribution:")
print(f"Total Positive = {pos_num}")
print(f"Total Negative = {neg_num}")
print(f"Total Neutral = {neu_num}")

# Text preprocessing
stop_words = stopwords.words('english')
processed_tweets = []

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters and URLs
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'https?://\S+', '', text)  # Remove URLs

    # Tokenize text
    tokens = word_tokenize(text)

    # Filter out stopwords
    filtered_tokens = [token for token in tokens if token not in stop_words]

    # Join tokens back into string
    return ' '.join(filtered_tokens)

# Process all tweets
for tweet in tweet_list:
    processed_tweets.append(preprocess_text(tweet))

# Feature extraction using Count Vectorizer
vectorizer = CountVectorizer(binary=False, min_df=2, ngram_range=(1, 2))
cv_result = vectorizer.fit_transform(processed_tweets)

# Display feature matrix dimensions
print("\nFeature Extraction Results:")
print(f"Dimension after Count Vectorizer: {cv_result.shape}")

# Determine optimal number of clusters using elbow method
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(cv_result)
    inertia.append(kmeans.inertia_)

# Plot elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal Cluster Selection')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()

# Based on the elbow plot, select appropriate number of clusters
optimal_clusters = 3  # This should be determined by examining the elbow plot

# Perform K-means clustering with optimal cluster count
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(cv_result)

# Add cluster information to our DataFrame
tweet_df['cluster'] = clusters

# Display cluster distribution
print(f"\nCluster Distribution (with {optimal_clusters} clusters):")
print(Counter(clusters))

# Print some example tweets from each cluster to understand what they represent
print("\nExample tweets from each cluster:")
for i in range(optimal_clusters):
    print(f"\nCluster {i}:")
    cluster_tweets = tweet_df[tweet_df['cluster'] == i]['text'].sample(5, random_state=42)
    for tweet in cluster_tweets:
        print(f"- {tweet[:100]}...")  # Show first 100 characters of each tweet
````

   - Clusters 1 and 2 reflect actual public feedback and grievances.

2. **PNR as Identifier**

   - Many tweets reference **PNR numbers**, showing that users want their issues tracked per journey.

3. **Service Issues Highlighted**

   - Cluster 2 includes **irrelevant or promotional tweets**, suggesting the need for better preprocessing or filtering before clustering.

The three clusters reveal different dimensions of communication between passengers and Indian Railways:

````python
# Step 1: Reshape the image data from (28, 28) to (784,)
x_train_reshaped = x_train.reshape(-1, 28*28)
x_test_reshaped = x_test.reshape(-1, 28*28)

print("New shape of x_train:", x_train_reshaped.shape)  # (60000, 784)
print("New shape of x_test:", x_test_reshaped.shape)    # (10000, 784)

# Step 2: Normalize pixel values to the range [0, 1]
x_train_normalized = x_train_reshaped / 255.0
x_test_normalized = x_test_reshaped / 255.0

# Step 3: Visualize some normalized images (reshape back to 28x28 for display)
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(x_train_normalized[i].reshape(28, 28), cmap='gray')
    plt.title(label_names[y_train[i]])
    plt.axis('off')
plt.suptitle("Normalized Fashion MNIST Images", fontsize=16)
plt.tight_layout()
plt.show()
````

````python
from sklearn.cluster import KMeans
import numpy as np

# Apply KMeans with 10 clusters
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
kmeans.fit(x_train_normalized)

# Predicted cluster labels for each image
cluster_labels = kmeans.labels_

# Create a DataFrame to map original and predicted clusters
import pandas as pd
kmeans_df = pd.DataFrame({
    'Original': y_train,
    'Predicted': cluster_labels
})

# Show number of images in each cluster
print("Cluster distribution:\n", kmeans_df['Predicted'].value_counts())

# Visualize 10 sample images from each cluster
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15))
for cluster_num in range(10):
    # Get indices of images in the current cluster
    cluster_indices = np.where(cluster_labels == cluster_num)[0][:10]

    for i, idx in enumerate(cluster_indices):
        plt.subplot(10, 10, cluster_num*10 + i + 1)
        plt.imshow(x_train_normalized[idx].reshape(28, 28), cmap='gray')
        plt.axis('off')
        if i == 0:
            plt.title(f"Cluster {cluster_num}", fontsize=10)
plt.suptitle("Cluster-wise Sample Images", fontsize=20)
plt.tight_layout()
plt.show()
````
