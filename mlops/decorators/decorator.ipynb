```python
def add_timing(func):
    """This function WRAPS another function with timing logic"""
    
    def wrapper():
        # Code to run BEFORE the original function
        print("üöÄ Starting function...")
        start = time.time()
        
        # Call the ORIGINAL function
        result = func()
        
        # Code to run AFTER the original function
        end = time.time()
        print(f"‚úÖ Function completed in {end - start:.2f}s")
        
        return result
    
    return wrapper  # Return the NEW function
```

```python
def simple_training():
    print("   Training model...")
    time.sleep(2)
    return "Model trained!"
```

```python
timed_training = add_timing(simple_training)
```

```python
result = timed_training()
print(f"Result: {result}")
```

```python
def mlops_timer(func):
    """A decorator for timing ML operations"""
    def wrapper(*args, **kwargs):
        func_name = func.__name__
        print(f"‚è±Ô∏è  Started {func_name}")
        
        start = time.time()
        result = func(*args, **kwargs)
        duration = time.time() - start
        
        # Smart formatting
        if duration < 1:
            duration_str = f"{duration*1000:.1f}ms"
        else:
            duration_str = f"{duration:.2f}s"
        
        print(f"‚úÖ {func_name} completed in {duration_str}")
        return result
    return wrapper

# Apply to ALL our ML functions
@mlops_timer
def load_data(filepath):
    print("   Loading data...")
    time.sleep(0.5)
    return "data_loaded"

@mlops_timer
def preprocess(data):
    print("   Preprocessing...")
    time.sleep(0.3)
    return "data_preprocessed"

@mlops_timer
def train_model(data):
    print("   Training model...")
    time.sleep(2)
    return "model_trained"

@mlops_timer
def evaluate(model):
    print("   Evaluating...")
    time.sleep(0.8)
    return {"accuracy": 0.95}

# Run a complete pipeline
data = load_data("train.csv")
clean_data = preprocess(data)
model = train_model(clean_data)
metrics = evaluate(model)
```

```python
def log_level(level):              # Level 1: Configuration
    print(f"üîß Level 1 executed with level='{level}'")
    
    def decorator(func):           # Level 2: Receives function
        print(f"üé® Level 2 executed with func={func.__name__}")
        
        def wrapper(*args, **kwargs):  # Level 3: Wraps execution
            print(f"üìù [{level}] Running {func.__name__}...")  # Uses 'level' from Level 1!
            result = func(*args, **kwargs)
            print(f"‚úÖ [{level}] {func.__name__} completed")
            return result
        
        return wrapper  # Return Level 3
    
    return decorator   # Return Level 2
```

```python
@log_level("DEBUG")
def train_model():
    print("   Training the model...")
    return "Model trained"

print("\n" + "="*50)
print("Now calling the function:")
print("="*50 + "\n")

result = train_model()
print(f"Result: {result}")
```

```python
import functools

def better_timer(func):
    """An improved timer decorator"""
    @functools.wraps(func)  # ‚Üê This is the magic!
    def wrapper(*args, **kwargs):
        print("Timing...")
        return func(*args, **kwargs)
    return wrapper

@better_timer
def train_model(epochs, learning_rate=0.01):
    """
    Trains a machine learning model.
    
    Args:
        epochs: Number of training epochs
        learning_rate: Learning rate for optimization
        
    Returns:
        Trained model
    """
    print(f"Training for {epochs} epochs with lr={learning_rate}")
    return "Model trained"

# Now let's inspect again
print("Function name:", train_model.__name__)
print("Function docstring:", train_model.__doc__)
print("\n" + "="*50)
help(train_model)
```

```python
import functools

# ‚ùå WITHOUT @functools.wraps
def bad_decorator(func):
    def wrapper(*args, **kwargs):
        """I'm the wrapper docstring"""
        return func(*args, **kwargs)
    return wrapper

# ‚úÖ WITH @functools.wraps
def good_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        """I'm the wrapper docstring"""
        return func(*args, **kwargs)
    return wrapper

def original_function(x, y):
    """I'm the original docstring"""
    return x + y

# Apply both decorators to copies of the function
bad_version = bad_decorator(original_function)
good_version = good_decorator(original_function)

# Compare
print("="*60)
print("WITHOUT @functools.wraps:")
print(f"  Name: {bad_version.__name__}")
print(f"  Doc: {bad_version.__doc__}")
print()
print("WITH @functools.wraps:")
print(f"  Name: {good_version.__name__}")
print(f"  Doc: {good_version.__doc__}")
print("="*60)
```

```python
import functools
import time
from datetime import datetime

def timer(func):
    """
    Time function execution with detailed logging
    """
    @functools.wraps(func)  # ‚Üê Preserves metadata of decorated functions
    def wrapper(*args, **kwargs):
        function_name = func.__name__  # ‚Üê This gets the REAL name, not 'wrapper'
        start_time = time.time()
        
        print(f"‚è±Ô∏è  Started {function_name} at {datetime.now().strftime('%H:%M:%S')}")
        
        try:
            result = func(*args, **kwargs)
            duration = time.time() - start_time
            
            if duration < 1:
                duration_str = f"{duration*1000:.1f}ms"
            elif duration < 60:
                duration_str = f"{duration:.2f}s"
            else:
                minutes = int(duration // 60)
                seconds = duration % 60
                duration_str = f"{minutes}m {seconds:.1f}s"
            
            print(f"‚úÖ {function_name} completed in {duration_str}")
            return result
            
        except Exception as e:
            duration = time.time() - start_time
            print(f"‚ùå {function_name} failed after {duration:.2f}s: {str(e)}")
            raise
    
    return wrapper

# Test it
@timer
def train_model(model_type, epochs=100):
    """Train an ML model with specified parameters"""
    time.sleep(1)
    return f"{model_type} trained for {epochs} epochs"

# This works correctly because of @functools.wraps
print("Function name:", train_model.__name__)  # Prints: train_model (not wrapper)
print("Docstring:", train_model.__doc__)       # Prints: Train an ML model...
help(train_model)  # Shows proper documentation

# Call it
result = train_model("RandomForest", epochs=50)
```
