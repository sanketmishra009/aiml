```python
x = np.arange(-5 , 5 , 0.225)
y = 1 / (1 + np.e**(-x))
grad = y * (1-y)
plt.plot(x , y)
plt.plot(x , grad)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate many independent random variables (uniform distribution)
n_samples = 100000
n_inputs = 100

# Inputs: uniform(-1,1)
X = np.random.uniform(-1, 1, size=(n_samples, n_inputs))

# Step 2: Linear combination (weights sampled from N(0,1))
weights = np.random.randn(n_inputs)
linear_combination = X.dot(weights)  # Wu

# Step 3: Apply nonlinearities
relu_output = np.maximum(0, linear_combination)      # ReLU
sigmoid_output = 1 / (1 + np.exp(-linear_combination))  # Sigmoid

# Plot distributions
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.hist(linear_combination, bins=80, density=True, color="steelblue", alpha=0.7)
plt.title("Linear Combination (Wu) \n ~ Gaussian-like by CLT")
plt.xlabel("Value")
plt.ylabel("Density")

plt.subplot(1,3,2)
plt.hist(relu_output, bins=80, density=True, color="darkorange", alpha=0.7)
plt.title("After ReLU (Nonlinear) \n Half-Normal, Skewed")
plt.xlabel("Value")

plt.subplot(1,3,3)
plt.hist(sigmoid_output, bins=80, density=True, color="seagreen", alpha=0.7)
plt.title("After Sigmoid (Nonlinear) \n Bounded, Skewed")
plt.xlabel("Value")

plt.tight_layout()
plt.show()
```

```python
x = np.arange(0 , 361 , 30)
sinx = np.sin((np.pi / 180)* x)
cosx = np.cos(np.radians(x))
plt.plot(x , sinx ,label = 'sinx')
plt.plot(x , cosx , label = 'cosx')
# plt.axvline(90 , linestyle = '--')
plt.axhline(0 , linestyle = '--')
plt.legend()
plt.show()
```
